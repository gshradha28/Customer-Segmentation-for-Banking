```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile) for 'Avg_Credit_Limit'
Q1 = data['Avg_Credit_Limit'].quantile(0.25)
Q3 = data['Avg_Credit_Limit'].quantile(0.75)

# Calculate the IQR
IQR = Q3 - Q1

# Define the upper threshold as Q3 + 1.5 * IQR
upper_threshold = Q3 + 1.5 * IQR

# Display the calculated threshold
upper_threshold

# Remove outliers beyond the upper threshold
filtered_data = data[data['Avg_Credit_Limit'] < upper_threshold]

# Display the first few rows of the filtered data
filtered_data.head()
```

```python
# Create a new column 'Credit_Category' based on Avg_Credit_Limit
data['Credit_Category'] = pd.cut(data['Avg_Credit_Limit'], 
                                 bins=[0, 20000, 50000, float('inf')],  # Using 'inf' to capture all high values
                                 labels=['Low', 'Medium', 'High'],
                                 include_lowest=True)  # Include the lower bound


print(data[['Avg_Credit_Limit', 'Credit_Category']].head())

```

```python
from sklearn.preprocessing import StandardScaler

# Select numerical columns for scaling
numerical_cols = ['Avg_Credit_Limit', 'Total_Credit_Cards', 
                  'Total_visits_bank', 'Total_visits_online', 'Total_calls_made']

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the numerical columns
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])

# Display the first few rows to verify the changes
print(data.head())
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Elbow Method to find the optimal number of clusters
sse = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data[numerical_cols])
    sse.append(kmeans.inertia_)

# Plot the Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(k_range, sse, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.show()



```

```python
# Assuming k = 3 (or the optimal value you found)
kmeans = KMeans(n_clusters=3, random_state=42)
data['Cluster'] = kmeans.fit_predict(data[numerical_cols])

# Display cluster centroids
print(kmeans.cluster_centers_)
```

```python
# Cluster profiling by finding the mean of each cluster
cluster_profile = data.groupby('Cluster')[numerical_cols].mean()
print(cluster_profile)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Perform Hierarchical Clustering
Z = linkage(data[numerical_cols], method='ward')

# Plot the Dendrogram
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster

# Choose the number of clusters
num_clusters = 3
data['HCluster'] = fcluster(Z, num_clusters, criterion='maxclust')

# Check cluster membership
print(data['HCluster'].value_counts())
```

```python
# Cluster profiling for hierarchical clusters
hcluster_profile = data.groupby('HCluster')[numerical_cols].mean()
print(hcluster_profile)
```

```python
# Compare the clusters from K-means and Hierarchical Clustering
comparison = data.groupby(['Cluster', 'HCluster']).size()
print(comparison)
```
